```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
df=pd.read_excel("Fuel_Consumption.xlsx")
```

```python
df.head()
```

```python
df.describe()
```

```python
cdf=df[["Engine_Size_L",
"Cylinders",
"Fuel_Consumption_City_L_per_100_km",
"CO2_Emissions_g_per_km"
]]
```

```python
cdf.head(9)
```

# Train/Test Splitting

```python
msk=np.random.rand(len(cdf)) <0.8
train=cdf[msk]
test=cdf[~msk]
```

```python
plt.scatter(train.Engine_Size_L , train.CO2_Emissions_g_per_km , color = 'blue')
plt.xlabel("Engine Size")
plt.ylabel("CO2 Emission")
plt.show()
```

# Model

```python
from sklearn import linear_model
regr = linear_model.LinearRegression()
x_train=np.asanyarray(train[["Engine_Size_L" , "Cylinders" , "Fuel_Consumption_City_L_per_100_km"]])
y_train=np.asanyarray(train[["CO2_Emissions_g_per_km"]])
regr.fit(x_train , y_train)
print("Coefficients" ,regr.coef_ )
```

#  Prediction

```python
x_test=np.asanyarray(test[["Engine_Size_L" , "Cylinders" , "Fuel_Consumption_City_L_per_100_km"]])
y_test=np.asanyarray(test[["CO2_Emissions_g_per_km"]])
y_hat_test=regr.predict(x_test)
print("MSE - " , np.mean((y_test - y_hat_test)**2))
print("Regression Score - " ,regr.score(x_test,y_test))
```
